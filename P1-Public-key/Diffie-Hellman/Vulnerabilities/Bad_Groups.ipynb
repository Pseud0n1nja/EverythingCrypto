# Prerequisites

- Group theory

# Using an additive group

$p$ = prime, $g$ = element with high order, $a, b$ = secret keys
- $A = ag \ \bmod \ p$
- $B = bg \ \bmod \ p$

The vulnerability is pretty obvious as we can easily calculate the inverses of $A$ and $B$ and multiply with them to find out $a, b$

# Subgroups

Observation: the discrete logarithm problem in a group $G$ is only as hard as the discrete logarithm problem in the largest subgroup of prime order in $G$ where $G$ is a finite abelian group

## Choosing a bad element to generate public keys

For this case suppose $|G|$ is not prime. Not every element in a group is a generator for that group => there are cyclic subgroups generated by various elements. The number of elements in these subgroups is smaller then then number of elements in the original group therefore limiting the posibilities of the public keys A and B.

Let $g$ be an element in the group $G$ => $g$ generates some subgroup $H = {g, g^2, g^3, ...}$ with $|H| < |G|$
- $A = g^a \ \bmod \ p \Rightarrow A \in H$
- $B = g^b \ \bmod \ p \Rightarrow B \in H$

From [Lagrange's theorem](https://en.wikipedia.org/wiki/Lagrange%27s_theorem_(group_theory)) we know  $|H|$ divides $|G|$.
If $|H|$ is small enough we can brute force an attack to find $a$ or $b$.

Let QR = Quadratic residue, NQR = non-quadratic residue

**A compromise**: Even if you're choosing the best $g$ you will make some compromise: 
- if $g$ is a QR => $|H| = |G| / 2$ because $g$ can only generate QR
- if $g$ is not a QR => you will leak the LSB of $a$ and $b$
    - if $A$ is QR => $A = g^{2y} => a = 2y => $ a is even
    - if $A$ is NQR => $a$ is odd
